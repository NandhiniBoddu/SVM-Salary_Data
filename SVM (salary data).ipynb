```python
import pandas as pd 
import numpy as np 
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.svm import SVC 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
train_data = pd.read_csv("SalaryData_Train(1).csv")
train_data 
```

```python
train_data=pd.get_dummies(train_data,columns=['workclass','education','maritalstatus','occupation','relationship','race','sex','native'])
train_data 
```

```python
label_encoder=preprocessing.LabelEncoder() 
```

```python
train_data['Salary']=label_encoder.fit_transform(train_data['Salary'])
train_data  
```

```python
X = pd.concat([train_data.iloc[:,0:5],train_data.iloc[:,6:103]],axis=1)
Y = train_data['Salary']  
```

```python
X 
```

```python
Y 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape 
```

# Grid Search CV

```python
clf = SVC
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
clf = SVC() #SVC Classification
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10) #model
gsv.fit(X_train,y_train) 
```
